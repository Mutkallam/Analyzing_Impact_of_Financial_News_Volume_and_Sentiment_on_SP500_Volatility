```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sp500 = pd.read_csv("sp500_raw_data.csv")
news = pd.read_csv("sp500_raw_headlines.csv")
```

```python
sp500.head()
```

```python
sp500.info()
```

```python
sp500["Date"] = pd.to_datetime(sp500["Date"])
sp500 = sp500.sort_values("Date")
sp500 = sp500[
    (sp500["Date"] >= "2010-01-01") &
    (sp500["Date"] <= "2023-12-31")
]
```

```python
(sp500.duplicated(subset=["Date"]).sum(),
 sp500.isna().sum())
```

```python
sp500["abs_return"] = abs((sp500["Close"] - sp500["Open"]) / sp500["Open"])
sp500["volatility"] = sp500["High"] - sp500["Low"]

sp500["volatility_next_day"] = sp500["volatility"].shift(-1)
sp500["abs_return_next_day"] = sp500["abs_return"].shift(-1)

sp500 = sp500.dropna()
```

```python
news.head()
```

```python
news = news[["Title", "Date"]]
```

```python
news.info()
```

```python
news = news.rename(columns={
    "Title": "Headline"})
```

```python
news["Date"] = pd.to_datetime(news["Date"])
news = news[
    (news["Date"] >= "2010-01-01") &
    (news["Date"] <= "2023-12-31")
]
```

```python
(news["Headline"].isna().sum(),
news.duplicated(subset=["Date", "Headline"]).sum())
```

```python
news = news.drop_duplicates(subset=["Date", "Headline"])
```

```python
sp500.to_csv("sp500_prepared.csv", index=False)
news.to_csv("news_prepared.csv", index=False)
```
